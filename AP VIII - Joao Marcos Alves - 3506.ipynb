```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
```

<H2> Importando o Dataframe
    

```python
path = "../Atividades/Atividade 8/weatherHistory.csv"
df = pd.read_csv(path)
df
```

```python
df["Summary"] = df["Summary"].astype('category').cat.codes
df["Precip Type"] =  df["Precip Type"].astype('category').cat.codes
df
```

<h2> Primeiramente, faça uma análise exploratória dos dados, entendendo as variáveis e suas
características. Crie gráficos par-a-par das variáveis para tentar identificar possíveis relações
lineares entre elas;

```python
colunas = df.columns.values.tolist()
colunas
```

```python
#sns.pairplot(df[["Temperature (C)",'Precip Type','Humidity',"Wind Speed (km/h)","Wind Bearing (degrees)","Pressure (millibars)"]])
sns.pairplot(df[colunas])
```

<p> Aparentemente, o único gráfico que apresenta uma aparência de regressão linear é o da temperatura, com a sensação térmica. Vamos usá-lo como análise.

```python
plt.scatter(df["Temperature (C)"],df["Apparent Temperature (C)"])
plt.xlabel("Temperatura atual")
plt.ylabel("Sensação térmica")
plt.title("Relação entre temperatura atual e Sensação térmica")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

x = df["Temperature (C)"]
y = df["Apparent Temperature (C)"]
lm = LinearRegression()

lm.fit(x.values.reshape(-1,1),y)
print("Coeficiente estimado: ",lm.coef_)
print("R2 (score): ",lm.score(x.values.reshape(-1,1),y))
```

<h2> Qual a equação linear e o coeficiente de determinação que representa a temperatura em termos da
umidade?

```python
x = df["Humidity"]
y = df["Temperature (C)"]
lm = LinearRegression()

lm.fit(x.values.reshape(-1,1),y)
print("Coeficiente estimado: ",lm.coef_)
print("R2 (score): ",lm.score(x.values.reshape(-1,1),y))
```

<h2> Qual a equação linear e o coeficiente de determinação que representa a temperatura em termos da
umidade e velocidade do vento?

```python
x = df[["Humidity","Wind Speed (km/h)"]]
y = df["Temperature (C)"]
lm = LinearRegression()

lm.fit(x,y)
print("Coeficiente estimado: ",lm.coef_)
print("R2 (score): ",lm.score(x,y))
```

<h2> Quais variáveis do dataset você indica para serem utilizadas para prever a umidade? Explique sua

```python
x = df.drop(['Humidity','Formatted Date','Summary','Daily Summary'],axis=1)
y = df["Humidity"]
lm = LinearRegression()
lm.fit(x,y)
print("Coeficiente estimado: ",lm.coef_)
print("R2 (score): ",lm.score(x,y))

coeff_df = pd.DataFrame(lm.coef_,x.columns,columns=['Coefficient'])
coeff_df
```
